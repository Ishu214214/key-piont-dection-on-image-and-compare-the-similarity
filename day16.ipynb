```python
matcher = cv2.BFMatcher()
```

```python
matches = matcher.knnMatch(des1, des2, k = 2) #kernal = k
len(matches)
```

```python
count = 0
th = 20
for m1, m2 in matches:
    gap = abs(m1.distance - m2.distance)
    if gap < th:
        count += 1
print(count, len(matches))
```
